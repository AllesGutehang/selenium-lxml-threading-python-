```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from lxml import etree
import time
import datetime
import openpyxl
import threading
```

## 加载网页

```python
def get_html(tar_url):
    #chrome_options = Options()
    #chrome_options.add_argument('--headless')
    driver=webdriver.Chrome(executable_path='G:/software/anaconda/chromedriver')#executable_path为chromedriver下载目录
    driver.get(tar_url)
    data=driver.page_source
    return data
```

## 用lxml解析

```python
#获取总览页面的主播基本信息
def get_zb_info(data,now_time):
    sum_list=[]
    element=etree.HTML(data)#使用lxml解析
    room_num=element.xpath('//li[@class="layout-Cover-item"]/div/a[1]/@href')#房间号
    r_name=element.xpath('//div[@class="DyListCover HeaderCell is-href"]//a[1]//h3[@class="DyListCover-intro"]/text()')#房间名
    zb_name=element.xpath('//li[@class="layout-Cover-item"]//a[1]//h2/text()')#主播名称
    game_name=element.xpath('//span[@class="DyListCover-zone"]/text()')#游戏名称
    hot=element.xpath('//div[@class="DyListCover-info"][2]//span[1]/text()')#热度
    for i in range(0,len(hot)):
        if room_num[i][0]=="/":#只保留直播的房间，去除放视频的房间
            sum_list.append([now_time,room_num[i],r_name[i],zb_name[i],game_name[i],hot[i]])#合并内容，同时插入当前时间now_time
        else:
            continue
    return sum_list
```

## 保存数据

```python
#将数据存入excel
def save_data(sum_list,r,excel):
    #r为起始行号，一般为2
    for field in sum_list:
        excel.cell(row=r,column=1,value=field[0])
        excel.cell(row=r,column=2,value=field[1])
        excel.cell(row=r,column=3,value=field[2])
        excel.cell(row=r,column=4,value=field[3])
        excel.cell(row=r,column=6,value=field[4])
        excel.cell(row=r,column=7,value=field[5])
        r+=1
    return r
```

## 定时方法(threading)

```python
def func(end_time_str):
    global timer#声明为全局变量
    n_time=datetime.datetime.now()
    end_time=datetime.datetime.strptime(end_time_str, "%Y-%m-%d %H:%M:%S")#将指定的结束时间装换成时间格式
    if n_time < end_time:
        #如果需要循环调用，就要添加以下方法
        print("hello")
        print(n_time)
        timer = threading.Timer(60, func,(end_time_str,))#参数解释（间隔时间（秒）,执行的函数,(函数的参数,))
        timer.start()
    else:
        timer.cancel()
```

```python
now_time = datetime.datetime.now()
tar_year=now_time.date().year
tar_month=now_time.date().month
tar_day=now_time.date().day
tar_time=datetime.datetime.strptime(str(tar_year)+"-"+str(tar_month)+"-"+str(tar_day)+" 13:30:00", "%Y-%m-%d %H:%M:%S")#设定你想要开始爬虫的时间
end_time_str=str(tar_year)+"-"+str(tar_month)+"-"+str(tar_day)+" 13:33:00"#设定你想要结束爬虫的时间
timer_start_time = (tar_time - now_time).total_seconds()#获取现在距离开始时间的间隔
print(timer_start_time)
timer_start = threading.Timer(timer_start_time, func,(end_time_str,))
timer_start.start()
```

## 定时循环爬虫

```python
#爬虫的函数
r=2
def run(r):
    tar_url='https://www.douyu.com/g_TVgame'
    target_source=r'G:\研二\斗鱼爬虫.xlsx'
    doc=openpyxl.load_workbook(target_source)
    excel=doc['斗鱼主机区']
    count=1
    #s_time=time.time()
    driver=webdriver.Chrome(executable_path='G:/software/anaconda/chromedriver')
    driver.get(tar_url)
    data=driver.page_source
    now_time=datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    result_list=get_zb_info(data,now_time)
    #print(r)
    r=save_data(result_list,r,excel)
    time.sleep(1)
    while driver.find_element_by_xpath('//span[text()="下一页"]/..').get_attribute('aria-disabled')=='false': 
        next=driver.find_element_by_xpath('//span[text()="下一页"]')
        next.click()
        time.sleep(1)
        data = driver.page_source
        #print(data)
        now_time=datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        result_list=get_zb_info(data,now_time)
        r=save_data(result_list,r,excel)
        count+=1
        #print(count)
    else:
        driver.quit()
        #print(r)
    doc.save(target_source)
    return r
```

```python
#定时循环的函数
def func(r,end_time_str):
    global timer
    n_time=datetime.datetime.now()
    end_time=datetime.datetime.strptime(end_time_str, "%Y-%m-%d %H:%M:%S")
    if n_time < end_time:
        #如果需要循环调用，就要添加以下方法
        r=run(r)
        timer = threading.Timer(1800, func,(r,end_time_str,))
        timer.start()
    else:
        timer.cancel()
```
